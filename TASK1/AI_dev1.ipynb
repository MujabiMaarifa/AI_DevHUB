```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

# read the  contents of the csv file
data = pd.read_csv('iris/iris.data')
print(data)
```

```python
# print the first five rows
new_data = data.head()
print(new_data)
```

```python
# check for null values
filter_data = new_data.info()
print(filter_data)
```

```python
# lets describe the new_data info
```

```python
summary = new_data.describe()
print(summary)
```

```python
# data visualization using the new data
new_data.plot(kind='hist')

plt.title("Histogram showing Iris Setosa Species over Frequency of occurence in the region")
plt.xlabel("Iris Setosa Species")
plt.show()
```

```python
# for the summary data
summary.plot(kind='bar')
plt.title("Bar graph showing distribution of Iris Setosa species against frequency")
plt.ylabel("Frequency")
plt.xlabel("Distribution")
plt.show()
```

```python
print(summary.columns)
```

```python
print(summary.index)  #checking for rows 
```

```python
# converting mean in rows to column
mean = summary.loc['mean']
count = summary.loc['count']

sns.scatterplot(x = mean, y=count)
plt.title("Scatter Diagram showing distribution of Count against mean")
plt.show()
```

```python
# scatter plot for the maximum value against minimum value
max = summary.loc['max']
min = summary.loc['min']

sns.scatterplot(x = min, y= max)
plt.title("Scatter diagram showing distribution of the min and max from summary")
plt.show()
```

```python
# boxplot showing the outliers of Iris setosa species 5.1
sns.boxplot(y=new_data['5.1'])
plt.title("Box plot for Iris Setosa species 5.1")
plt.ylabel("Frequency")
plt.xlabel("Distribution for Iris Setosa 5.1")
plt.show()
```
